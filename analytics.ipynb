```python
list2 = ['Date/Time', 'SO2', 'NO2','O3', 'CO', 'PM10', 'PM2.5', 'Lower Ambient Temperature', 'Upper Ambient Temperature','Ambient Temperature', 'Barometric Pressure', 'Relative Humidity','RelativeHumidity', 'Wind Direction', 'Wind Speed', 'H2S', 'Toluene', 'O-Xylene', 'Ethylbenzene', 'MP-Xylene', 'Benzene', 'CH4', 'NMHC', 'THC', 'Noise', 'Solar Radiation','Atm P.']
```

```python
def count_matching_values(sheet_name,list1, list2):
    matching_values = set([item.lower() for item in list1]) & set([item.lower() for item in list2])
    matching_count = len(matching_values)
    return sheet_name, matching_values, matching_count
```

```python
excel_file = pd.ExcelFile("/workspaces/vikrant_dubai/RAW/AJ/AJ 2.xlsx")

aic=[]

for i in range(len(excel_file.sheet_names)):
    df=pd.read_excel("/workspaces/vikrant_dubai/RAW/AJ/AJ 2.xlsx", sheet_name=excel_file.sheet_names[i])
    df = df.iloc[10:]
    df.columns = df.iloc[0]
    df = df[1:]
    list1 = list(df.columns)
    matching_count = count_matching_values(sheet_name=excel_file.sheet_names[i], list1=list1, list2=list2)
    aic.append(matching_count)
```
